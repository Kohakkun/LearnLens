```python
model = YOLO('./runs/detect/train4/weights/last.pt')
```

```python
img = os.path.join('data', 'images', 'Sleepy.0f84dba0-6508-11ef-8bb1-b03cdc8203d8.jpg')
```

```python
results = model(img)
for result in results:
    print("Detections:")
    for box in result.boxes: 
        print(f"Class: {box.cls}, Confidence: {box.conf}, Box Coordinates: {box.xyxy}")

    result.show()

for result in results:
    img_with_detections = result.plot()

    plt.imshow(img_with_detections)
    plt.axis('off')
    plt.show()
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        break

    # Make detections
    results = model(frame)  # 'results' is a list of results

    # Iterate over each result to render and visualize
    for result in results:
        # Render the detection on the image
        img_with_detections = result.plot()  # Returns the image with drawn detections
        
        # Display the image with detections
        cv2.imshow('YOLO', img_with_detections)
    
    # Exit condition
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

```

```python
# Release resources
cap.release()
cv2.destroyAllWindows()
```
